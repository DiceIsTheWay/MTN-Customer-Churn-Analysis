```python
# Create a data context

from great_expectations.data_context import FileDataContext

context = FileDataContext.create(project_root_dir='./')
```

```python
# Give a name to a Datasource. This name must be unique between Datasources.
datasource_name = 'csv-data-mtn'
datasource = context.sources.add_pandas(datasource_name)

# Give a name to a data asset
asset_name = 'mtn-cust-q1'
path_to_data = '/content/P2M3_Dais_data_clean.csv'
asset = datasource.add_csv_asset(asset_name, filepath_or_buffer=path_to_data)

# Build batch request
batch_request = asset.build_batch_request()
```

```python
# Creat an expectation suite
expectation_suite_name = 'expectation-mtn-dataset'
context.add_or_update_expectation_suite(expectation_suite_name)

# Create a validator using above expectation suite
validator = context.get_validator(
    batch_request = batch_request,
    expectation_suite_name = expectation_suite_name
)

# Check the validator
validator.head()
```

## Expectation

### 1. To be Unique

```python
# Expect column 'customer_id' to be unique
validator.expect_column_values_to_be_unique("customer_id")
```

### 2. To be Between min_value and max_value

```python
# Expect columns to be between their min_value and max_value
validator.expect_column_values_to_be_between("age", min_value=16, max_value=100)
validator.expect_column_values_to_be_between("satisfaction_rate", min_value=1, max_value=5)
validator.expect_column_values_to_be_between("customer_tenure_in_months", min_value=1, max_value=100)
validator.expect_column_values_to_be_between("unit_price", min_value=0, max_value=150000)
validator.expect_column_values_to_be_between("number_of_times_purchased", min_value=1, max_value=20)
validator.expect_column_values_to_be_between("total_revenue", min_value=250, max_value=3_000_000)
validator.expect_column_values_to_be_between("data_usage", min_value=0, max_value=200)
```

### 3. To be in Set

```python
# Expect columns to be in set
validator.expect_column_values_to_be_in_set("gender", ["Male", "Female"])
validator.expect_column_values_to_be_in_set("customer_review", ["Poor", "Fair", "Good", "Very Good", "Excellent"])
validator.expect_column_values_to_be_in_set("customer_churn_status", ["Yes", "No"])
```

### 4. To be in Type list

```python
# Expect column `data_usage` must in form of integer or float

validator.expect_column_values_to_be_in_type_list('data_usage', ['integer', 'float'])
```

### 5. Column to exist

```python
# Expect column 'customer_churn_status' to exist

validator.expect_column_to_exist(column='customer_churn_status')
```

### 6. Column count to be equal





```python
# Expect column count to be equal to 17
validator.expect_table_column_count_to_equal(value=17)
```

### 7. To be Table row count to be between

```python
# Table row count to be between 900 to 1000
validator.expect_table_row_count_to_be_between(min_value=900, max_value=1100)
```

```python
# Save into Expectation Suite

validator.save_expectation_suite(discard_failed_expectations=False)
```

## Checkpoint

```python
# Create a checkpoint

checkpoint_1 = context.add_or_update_checkpoint(
    name = 'checkpoint_1',
    validator = validator,
)
```

```python
# Run a checkpoint

checkpoint_result = checkpoint_1.run()
```
